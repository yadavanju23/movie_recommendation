```python
import ast
```

```python
def genre_extract(col):
    L=[]
    for i in ast.literal_eval(col):
        L.append(i['name'])
    return L
```

```python
data['genres']=data['genres'].apply(genre_extract)
```

```python
data.sample(3)
```

```python
data['keywords'][0]
```

```python
def keyword(col):
    L=[]
    for i in ast.literal_eval(col):
        L.append(i['name'])
    return L
```

```python
data['keywords']=data['keywords'].apply(keyword)
```
